```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from MyTools import my_line_graph
```

```python
from Platonic_Graphs import tetrahedron, pos_tetrahedron

LG_Tetrahedron, pos_LG_Tetrahedron = my_line_graph(tetrahedron, pos_tetrahedron)

nx.draw(tetrahedron, pos=pos_tetrahedron, node_color='blue',
        edge_color='blue', width=5)

nx.draw(LG_Tetrahedron, pos=pos_LG_Tetrahedron, node_color='red',
       edge_color='red', width=1)
```

```python
from Platonic_Graphs import cube, pos_cube, octahedron, pos_octahedron

## Constructing the line graphs of the cube/octahedron ##
LG_cube, pos_LG_cube = my_line_graph(cube, pos=pos_cube)
LG_octahedron, pos_LG_octahedron = my_line_graph(octahedron, pos=pos_octahedron)

## Plotting ##
fig, ax = plt.subplots(ncols=2,figsize=(20,10))
nx.draw(cube, ax=ax[0], pos=pos_cube, node_color='blue',
        edge_color='blue', width=1)

nx.draw(LG_cube, ax=ax[0], pos=pos_LG_cube, node_color='red',
        edge_color='red', width=5, with_labels=True)

nx.draw(octahedron, ax=ax[1], pos=pos_octahedron, node_color='blue',
        edge_color='blue', width=1)

nx.draw(LG_octahedron, ax=ax[1], pos=pos_LG_octahedron, node_color='red',
        edge_color='black', width=5, with_labels=True)
```

```python
order = [1,0,4,3,5,11,6,8,2,7,9,10]
transformed_pos_LG_octahedron = dict(zip(order, pos_LG_cube.values()))

transformed_pos_LG_octahedron[0] = (-2.0, 1.5)
transformed_pos_LG_octahedron[4] = (1.5, -2.0)
transformed_pos_LG_octahedron[5] = (1.5, 5.0)
transformed_pos_LG_octahedron[6] = (5.0, 1.5)

pos_LG_cube[1] = (-2.0, 1.5)
pos_LG_cube[2] = (1.5, -2.0)
pos_LG_cube[4] = (1.5, 5.0)
pos_LG_cube[6] = (5.0, 1.5)

### Plotting ###
fig, ax = plt.subplots(ncols=2,figsize=(20,10))

## Line graph of a cube of the left ##
nx.draw(LG_cube, ax=ax[0], pos=pos_LG_cube,
        node_color='red', edge_color='red', width=3, with_labels=True)

## Line graph of an octahedron on the right ##
nx.draw(LG_octahedron, ax=ax[1], pos=transformed_pos_LG_octahedron,
        node_color='red', edge_color='black', width=3, with_labels=True)
```

```python
cross_edges = [(7,9), (2,10), (0,6), (4,5), (1,3), (0,2), (1,8), (4,7), (6,10), (8,11), (5,9), (3,11)]
LG_Cube_edges = [e for e in LG_octahedron.edges if e not in cross_edges]

fig, ax = plt.subplots(ncols=3, figsize=(30,10))

## Plot 1- Far left ##
nx.draw_networkx_nodes(LG_octahedron, ax=ax[0], pos=transformed_pos_LG_octahedron)
nx.draw_networkx_edges(LG_octahedron, ax=ax[0], pos=transformed_pos_LG_octahedron,
                       edgelist=cross_edges, edge_color='black')
nx.draw_networkx_edges(LG_octahedron, ax=ax[0], pos=transformed_pos_LG_octahedron,
                       edgelist=LG_Cube_edges, edge_color='red')

## Plot 2 - Middle 
nx.draw_networkx_nodes(LG_octahedron, ax=ax[1], pos=transformed_pos_LG_octahedron)
nx.draw_networkx_edges(LG_octahedron, ax=ax[1], pos=transformed_pos_LG_octahedron,
                       edgelist=LG_Cube_edges, edge_color='red')

##
nx.draw_networkx_nodes(LG_octahedron, ax=ax[2], pos=transformed_pos_LG_octahedron)
nx.draw_networkx_edges(LG_octahedron, ax=ax[2], pos=transformed_pos_LG_octahedron,
                       edgelist=cross_edges, edge_color='black')
plt.show()
```

```python
from Platonic_Graphs import dodecahedron, pos_dodecahedron, icosahedron, pos_icosahedron

LG_Dodecahedron, pos_LG_Dodecahedron = my_line_graph(dodecahedron, pos_dodecahedron)
LG_Icosahedron, pos_LG_Icosahedron = my_line_graph(icosahedron, pos_icosahedron)

fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20,10))
# nx.draw(dodecahedron, ax=ax[0], pos = pos_dodecahedron,
#         node_color = 'blue', edge_color = 'blue', width=1)

nx.draw(LG_Dodecahedron, ax=ax[0], pos = pos_LG_Dodecahedron,
        node_color = 'red', edge_color = 'red', width=2, with_labels=True)

# nx.draw(icosahedron, ax=ax[1], pos = pos_icosahedron,
#         node_color = 'blue', edge_color = 'blue', width=1)

nx.draw(LG_Icosahedron, ax=ax[1], pos = pos_LG_Icosahedron,
        node_color = 'red', edge_color = 'black', width=2, with_labels=True)
```

```python
outside_pent = [3,0,1,7,5]
factor = 2.0
for i in outside_pent:
    x,y = pos_LG_Dodecahedron[i]
    newx = factor*x
    newy = factor*y
    pos_LG_Dodecahedron[i] = (newx,newy)
    
order = [7,24,6,26,8,14,15,23,13,22,2,0,9,5,27,16,12,17,21,25,18,20,3,1,10,19,28,29,4,11]
transformed_pos_LG_Icosahedron = dict(zip(order,pos_LG_Dodecahedron.values()))
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20,10))

nx.draw(LG_Dodecahedron, ax=ax[0], pos = pos_LG_Dodecahedron,
        node_color = 'red', edge_color = 'red', node_size=200,
        width=1, with_labels=True)

nx.draw(LG_Icosahedron, ax=ax[1], pos = transformed_pos_LG_Icosahedron,
        node_color = 'red', edge_color = 'black', node_size=200,
        width=1, with_labels=True)
```
